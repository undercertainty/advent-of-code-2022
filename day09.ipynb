```python
def get_path(str_in):
    o=[]
    for s in str_in.splitlines():
        (m, n)=s.split()
        o+=[m]*int(n)
    
    return o
```

Put into a function:

```python
def day09_b(str_in):

    states_ls=[(0, 0, 'X') for i in range(10)]
    tail_path={(0, 0)}


    for d in get_path(str_in):
        next_move=d
        for (i, state) in enumerate(states_ls):
            (states_ls[i], next_move)=move_chain(states_ls[i], next_move)
        tail_path.add((states_ls[-1][0], states_ls[-1][1]))


    return len(tail_path)
```
